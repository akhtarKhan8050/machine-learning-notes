```python
import pandas  as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('Datasets\creditcard.csv')
data
```

```python
data.info()
```

```python
data['Class'].value_counts()
# To handle imbalnace i have a technique called SMOTE (Synthetic Minority Oversampling Technique)
# to methods in smote oversampling and undersampling

# It takes a sample from minority class
# knn finds the nearest neighbours from the same class
# pick one of the neighbours randomly
# it creates a new synthetic sample somewhere between the originak and and ts neighbour
# Oversampling and undersampling
```

```python
scaler=StandardScaler()
#Stnaard scaer model works on 2d data that is rows and columns so to access one column from the dataframe we use data[[column_name]]
data['normAmount'] = scaler.fit_transform(data[['Amount']])
data = data.drop(['Time', 'Amount'], axis=1)
data.head()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X=data.drop(['Class'],axis=1)
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))
```

```python
sm=SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())
```

```python
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix #To load the metrcs involved for prediction evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import seaborn as sns
cm=confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print("Recall metric in the testing dataset: {}%".format(100*cm[1,1]/(cm[1,0]+cm[1,1])))
```
